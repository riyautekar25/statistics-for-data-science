# What is descriptive statistics?
As the name suggest it describes features of your data such as centarl tendancy (mean, median, mode) and variability (range, standard deviation and variance).

# Purpose:
1. Summarize large datasets
2. Provide foundation for more complex statistical anaysis (inferential statistics, wil cover this later)

# TOPIC 1:
## CENTRAL TENDENCY
A single center value of your data that represents or summarizes entire dataset.
1. Mean: Average value of your data : When to use: Numerical data + roughly symmetrical data + NO extreme outliers --> NOTE** MEAN is affected by outliers
2. Median: Middle value when data is sorted: When to use: Numerical data + outliers are present + Data is skewed + you want typical value --> Median is robust to outliers
3. Mode: Most frequent value in your data: When to use: Categorical Data (can be used in Numerical case also) + you want most common cases

```python
import pandas as pd 

data = {
    "exam_scores":[45, 56, 67, 45, 78, 89, 90, 56, 67, 45]
}

df = pd.DataFrame(data)
df
```

```python
mean_score = df["exam_scores"].mean()
median_score = df["exam_scores"].median()
mode_score = df["exam_scores"].mode()[0]   #here we usedmode()[0]--> this will return first mode, your data can be multimodel at times(may be bimodal), and in sucha case to return only one mode answer we use [0]

print(mean_score, median_score, mode_score)
```

```python
import numpy as np

# data_numpy = np.array([1, 2, 3, 4, 4, 5, 5, 5, 6])
data_numpy = df["exam_scores"] # here df['exam_scores'] is of type series which is nothing but an numoy array

numpy_mean = np.mean(data_numpy)
numpy_median = np.median(data_numpy)
#NumPy does not have a built-in mode() function

numpy_mean, numpy_median
```

USING statistics Module (Standard Library)

```python
import statistics

# data_stat = np.array([1,1,2,3,4,5,5,6])
data_stat = df["exam_scores"]
stat_mean = statistics.mean(data_stat)
stat_median = statistics.median(data_stat)
stat_mode = statistics.mode(data_stat)

# for multimodal data
stat_multi_mode = statistics.multimode([1,1,2,2,3])
print(stat_mean, stat_median, stat_mode, stat_multi_mode)
```

```python
import matplotlib.pyplot as plt

plt.hist(df["exam_score"], bins=5)
plt.axvline(mean_score)
plt.axvline(median_score)
plt.title("Distribution of Exam Scores")
plt.show()
```
